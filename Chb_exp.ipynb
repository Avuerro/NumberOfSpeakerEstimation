```python
def normalize_sample(sample, output):
    print(sample)
    command = f"ffmpeg-normalize {sample} --target-level 0 -o {output}"
    subprocess.call(command, shell=True)

normalize_sample(sample_1_loc,'./samples/normalized-84-121123-0000.wav')
```
